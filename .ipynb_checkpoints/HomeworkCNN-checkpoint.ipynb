## Data inlezen en voorbereiden

```python
data_dir = pathlib.Path(data_filepath)
```

Het aantal afbeeldingen dat zal worden gebruikt wordt hieronder nog eens uitgeprint. In de mappen kan er ook een `.DS_Store` bestand staan dat hieronder in de optelling niet mag worden meegenomen.

```python
image_count = len([s for s in list(data_dir.glob('*/*.*')) if s.name != ".DS_Store"])
print("Aantal afbeeldingen:", image_count)
```

De **train_ds** (trainingsdataset) en de **val_ds** (testdataset) worden hieronder met de keras utility gesplitst in 2 delen. De variabele bevat de labels en de afbeeldingen. De datasets worden tijdens het opspitsen geshuffled om op een willekeurige manier te worden verdeeld.

Er wordt gekozen voor een afbeeldingsgrootte van 128 bij 128. De batchsize heb ik ingesteld op 32 omdat het een standaard waarde is dat meestal gebruikt wordt in kleine datasets (+/- 1000 samples). Het vergroten of verkleinen zal geen groot verschil maken voor het probleem, tenzij er wordt getraind op honderdduizenden of miljoenen samples.

De **image_dataset_from_directory** functie zorgt automatisch op een eenvouige manier voor het inlezen van de afbeeldingen volgens de mappenstructuur.

```python
BATCH_SIZE = 32
img_height = 128
img_width = 128

train_ds = tf.keras.utils.image_dataset_from_directory(
  data_filepath,
  validation_split=0.25,
  subset="training",
  shuffle=True,
  seed=123,
  image_size=(img_height, img_width),
  batch_size=BATCH_SIZE)
```

```python
val_ds = tf.keras.utils.image_dataset_from_directory(
  data_filepath,
  validation_split=0.25,
  subset="validation",
  shuffle=True,
  seed=123,
  image_size=(img_height, img_width),
  batch_size=BATCH_SIZE)
```

Er zijn in dit geval 5 soorten vlinders die hieronder in een object worden bijgehouden. Omdat train en test -dataset de klasses bijhoudt volgens een integer waarde is het eenvoudiger om een klasse te vinden volgens key in de onderstaande `class_names` variabele.

```python
class_names = {i : train_ds.class_names[i] for i in range(0, len(train_ds.class_names))}
class_names
```

De train en test -dataset bevatten beide 32 samples van 128 bij 128 met telkens 3 pixel (RGB) waarden. Voor elke sample is er ook een label voorzien.

```python
for train_images, train_labels in train_ds:
    print("train images shape:", train_images.shape)
    print("train labels shape:", train_labels.shape)
    break
    
for test_images, test_labels in val_ds:
    print("test images shape:", test_images.shape)
    print("test labels shape:", test_labels.shape)
    break
```

## Model definitie

In deze sectie stel je het neural netwerk samen en wordt het getraind op de trainingsdata. Gebruik <code>model</code> als variabelenaam voor je model.

Het model start met het aanmaken van een Sequential model. In de array parameter worden de volgende zaken toegevoegd:

- Rescaling laag
- Convolutional 2D laag
- MaxPooling 2D laag
- Flatten laag
- Dropout laag
- Dense laag

### Rescaling laag
De herschalings laag herschaalt elke waarde van een input (vaak een afbeelding) door te vermenigvuldigen met een schaal factor en er een offset toe te voegen. Het schalen van afbeeldingen wordt **image normalization** genoemd. Dit proces is handig om een uniforme schaal te bieden voor de dataset of numerieke waarden die gebruikt worden voordat het model er verder mee aan de slag gaat.

### Convolutional 2D laag
Een Convolutionele laag is een lineaire bewerking waarbij een reeks gewichten met de invoer wordt vermenigvuldigd. Aangezien de techniek is ontworpen voor tweedimensionale invoer, wordt de vermenigvuldiging uitgevoerd tussen een reeks invoergegevens en een tweedimensionale reeks gewichten, ook wel een filter of een kernel genoemd.

Als eerste parameter worden het aantal `filters` (neuronen) opgegeven, dit vertelt de laag hoeveel inputs er worden aangenomen om op te trainen.

De 2de parameter bevat een tuple waarin de `kernel_size` wordt opgegeven dat de breedte en hoogte van het 2D-convolutievenster bevat. Typisch waarden voor de kernel_size zijn: `(1, 1), (3, 3), (5, 5), (7, 7`. Het is zeldzaam om kernelgroottes groter dan 77 te gebruiken.

Als 3de parameter wordt de activatie functie opgenomen in dit geval de ReLU activatie functie voor alle lagen (behalve de laatste).

Als 4de parameter (enkel bij de eerste input laag) wordt de input shape opgegeven dat gelijk moet zijn aan de grootte van de geschaalde afbeeldingen in de trainings dataset.

#### ReLU
De **rectified linear activation** functie of `ReLU` in het kort, is een lineare functie dat de input teruggestuurd wanneer het een positief getal is, in het andere geval wordt er 0 teruggestuurd.

#### SoftMax
De output van de `softmax` activatie functie wordt toegewezen aan een 0 tot 1 bereik. Het brengt de output in kaart op een manier dat de totale som van alle waaren gelijk is aan 1. Op deze manier worden de probabiliteiten voor elke eindklasse aangegeven. Voor een clasificatie probleem wordt de eindlaag ingesteld op een softmax activatie functie met het aantal filters ingesteld op het aantal klassen.

### MaxPooling 2D laag
`Max pooling` verkleint de invoer langs zijn ruimtelijke afmetingen (hoogte en breedte) door de maximale waarde te nemen over een bepaalt inputvenster (gedefinieerd door de pool_size parameter) voor elk kanaal van de invoer. Het venster wordt stapsgewijs langs elke dimensie verschoven om zo tot een kleinere output te komen.

### Flatten laag

De `Flatten` laag wordt gebruikt om de multidimensionale invoer eendimensionaal te maken, vaak wordt deze laag gebruikt bij de overgang van de convolutielaag naar de fully connected laag (output laag).

### Dropout laag
De `Dropout` laag zorgt ervoor dat tijdens de training er een aantal neuronen willekeurig genegeerd worden. Dit heeft als effect dat de laag wordt behandeld met een minder aantal knooppunten en connectiviteit met de vorige laag. Dit heeft tot gevolg dat het trainingsproces luidruchtig wordt, waardoor knooppunten binnen een laag worden gedwongen om waarschijnlijk meer of minder verantwoordelijkheid voor de invoer op zich te nemen. Vaak wordt er maar 20% uitgeschakeld zodat het niet te veel nodes uitschakeld.

### Dense laag
een `Dense` laag is een laag dat diep verbonden is met de voorgaande laag, wat betekent dat de neuronen van de laag verbonden zijn met elk neuron van de voorgaande laag. Elk neuron van de Dense laag doet als basis functie matrix-vectorvermenigvuldiging. Matrixvectorvermenigvuldiging is een procedure waarbij de rijvector van de uitvoer van de voorgaande lagen gelijk is aan de kolomvector van de Dense laag.


##### Bronnen

[Link naar keras-conv2d-and-convolutional-layers](https://www.pyimagesearch.com/2018/12/31/keras-conv2d-and-convolutional-layers/) 

[Link naar a-complete-understanding-of-dense-layers-in-neural-networks](https://analyticsindiamag.com/a-complete-understanding-of-dense-layers-in-neural-networks/)

[Link naar convolutional-layers-for-deep-learning-neural-networks](https://machinelearningmastery.com/convolutional-layers-for-deep-learning-neural-networks/)

```python
model = tf.keras.Sequential([
  tf.keras.layers.Rescaling(1./255),
  tf.keras.layers.Conv2D(192, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)),
  tf.keras.layers.MaxPooling2D(pool_size=(3, 3)),
  tf.keras.layers.Conv2D(256, (3, 3), activation='relu'),
  tf.keras.layers.MaxPooling2D(pool_size=(3, 3)),
  tf.keras.layers.Conv2D(128, (2, 2), activation='relu'),
  tf.keras.layers.MaxPooling2D(pool_size=(3, 3)),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dense(5, activation='softmax')
])
```

### Compileren van het model

Nadat het model gedefineerd is wordt het model gecompileerd met enkele parameters.

#### Optimizer
Als optimizer wordt er gekozen voor de `adam` optimizer. De adam optmizer gebruikt Stochastic gradient descent dat 1 learning rate (termed alpha) gebruikt voor alle gewichten, de learning rate veranderd ook niet tijdens het trainen.

het leertempo (learning rate) wordt bijgehouden voor elk gewicht (parameter) in het netwerk en afzonderlijk aangepast tijdens het trainingsproces.

#### Loss
SparseCategoricalCrossentropy en CategoricalCrossentropy berekenen beide categorische cross-entropy. Het enige verschil is hoe de targets/labels moeten worden gecodeerd.

Bij de `SparseCategoricalCrossentropy` worden de doelen weergegeven door de index van de categorie (vanaf 0) te nemen. Het herhaalt alle mogelijke klassen C die door het model zijn voorspeld tijdens de voorwaartse doorloop van het trainingsproces. Voor elke klasse wordt gekeken naar de doelobservatie van de klas, d.w.z. of de werkelijke klas die overeenkomt met de voorspelling in de trainingsset 0 of 1 is. Hiernaast berekent het de logaritme van de voorspelling van de waarneming (de kans dat het tot die klasse behoort). Hieruit volgt dat slechts n zo een waarde relevant is: het eigenlijke doel. Hiervoor berekent het eenvoudig de natuurlijke logwaarde die aanzienlijk toeneemt wanneer deze verder weg is van 1.

#### Metrics
De metriek parameter neemt een array aan waarin bepaald wordt welke functies er worden gebruikt om de prestaties van het model te beoordelen. Meestal (ook in dit geval) wordt de `Accuracy` gebruikt om dit te bepalen. 


```python
model.compile(
    optimizer='adam',
    loss=tf.losses.SparseCategoricalCrossentropy(),
    metrics=['accuracy']
)
```

Na het compileren wordt het model gefit met als inputs de trainingsdataset met afbeeldingen en labels, het aantal epochs (het aantal keer de gehele dataset wordt overlopen), de testdataset (validatie data) met afbeeldingen en labels en de batchsize (het aantal afbeelingen dat per iteratie wordt gebruikt) meegegeven. Het aantal epochs staat op 7 omdat na 10 epochs het trainen op het maximum komt er duidelijke overfitting begint op te treden.

```python
NUM_EPOCHS = 7
BATCH_SIZE = 32
```

```python
history = model.fit(train_ds, epochs=NUM_EPOCHS, validation_data=val_ds, batch_size=BATCH_SIZE)
```

Hieronder worden alle lagen met het aantal parameters weergegeven. Het totaal aantal parameters waarop het model heeft getrained is 710,661.

De loss is een scalaire waarde die we proberen te minimaliseren tijdens het trainingproces van het model. Hoe lager het verlies, hoe dichter onze voorspellingen bij de echte labels liggen.

Na het trainen krijgt de evaluatie een loss en accuracy waarde terug dat een `loss` heeft van 0.5561 en een `accuracy` van 81.25%. Volgen deze evaluatie scoort het model best goed op de test data maar is er een zekere vorm van overfitting.

### Model testen

Om een model te testen op ongeziene data kan er aan de onderstaande functie een model en een url van een afbeelding worden meegegeven. De afbeelding wordt vervolgens ingeladen, omgezet en aan het model gegeven. Hierna test het model de afbeelding en geeft het een probabiliteit score aan de 5 klasses. De hoogste score wordt geprint met de klasse naam dat het voorspelde. Om dit te testen heb ik 5 url's van de 5 soorten vlinders opgzocht en in een object gestopt. Hierdoor kan het model op een eenvoudige manier getest worden door de index op te geven aan het object.

```python
def test_butterly_prediction(model, butterfly_url):
    butterfly_path = tf.keras.utils.get_file(origin=butterfly_url)

    img = tf.keras.utils.load_img(butterfly_path, target_size=(img_height, img_width))
    img_array = tf.keras.utils.img_to_array(img)
    img_array = tf.expand_dims(img_array, 0)

    butterfly_predictions = model.predict(img_array)
    butterfly_score = tf.nn.softmax(butterfly_predictions[0])

    print("Probabiliteiten over alle 5 soorten", butterfly_predictions)
    print()
    print("Deze afbeelding is met {:.2f}% zekerheid een {} vlinder."
        .format(100 * np.max(butterfly_predictions[0]), class_names[np.argmax(butterfly_score)]))
    return (img)

butterfly_list = {
     0: "https://pennybrownillustrations.files.wordpress.com/2016/10/bhutan-glory-resized.jpg",
     1: "https://upload.wikimedia.org/wikipedia/commons/a/ae/Ornithoptera_chimaera_m.jpg",
     2: "https://upload.wikimedia.org/wikipedia/commons/0/0f/BinghamTeinopalpus.jpg",
     3: "https://upload.wikimedia.org/wikipedia/commons/e/ec/Apodemia_mormo_langei-02.JPG",
     4: "https://upload.wikimedia.org/wikipedia/en/5/56/PVBlue.jpg"}
```

```python
#butterfly_list gebruikt de zelfe index als class_names
class_names
```

```python
img = test_butterly_prediction(model, butterfly_list[2])
print("afbeelding:")
img
```

## Data augmentatie

### ImageDataGenerator

De ImageDataGenerator zorgt ervoor dat het model nieuwe variaties krijgt van dezelfde afbeelding elke epoch. Het  returned enkel de getransformeerde afbeelding en voegt het niet toe aan de originele dataset. Als het wel zou worden toegevoegd aan het model, dat zou het model overfitten op de afbeeldingen omdat het dezelfde afbeelding meerdere keren te zien krijgt.

De onderstaande instellingen zorgen ervoor dat de afbeelding tot 45 zal gedraaid wordt, de afbeeling tot 20% wordt ingezoomd en dat het mogelijks horizontaal omgedraaid wordt.

Het Sequential model gaat ook zorgen dat dit gebeurd maar de ImageDataGenerator wordt toegepast op de fit als parameter met de .flow methode.

```python
data_gen = ImageDataGenerator(
    rotation_range=0.45,  
    zoom_range = 0.2, 
    horizontal_flip=True,
    fill_mode="nearest"
)
```

Het geaugmenteerde model wordt zoals het oorspronkelijke model opgebouwd met dezelfde lagen.

```python
augmentation_model = Sequential([
  tf.keras.layers.Rescaling(1./255),
  tf.keras.layers.Conv2D(192, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)),
  tf.keras.layers.MaxPooling2D(pool_size=(3, 3)),
  tf.keras.layers.Conv2D(256, (3, 3), activation='relu'),
  tf.keras.layers.MaxPooling2D(pool_size=(3, 3)),
  tf.keras.layers.Conv2D(128, (4, 4), activation='relu'),
  tf.keras.layers.MaxPooling2D(pool_size=(3, 3)),
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(64, activation='relu'),
  tf.keras.layers.Dense(5, activation='softmax')
])
```

```python
augmentation_model.compile(
    optimizer='adam',
    loss=tf.losses.SparseCategoricalCrossentropy(),
    metrics=['accuracy']
)
```

```python
augmentation_history = augmentation_model.fit(data_gen.flow(train_images, train_labels,
                                    batch_size=BATCH_SIZE,seed=27, shuffle=False), 
                                    validation_data=val_ds, epochs=NUM_EPOCHS)
```

```python
class_names
```

```python
butterfly_list
```

Uit onderstaande tests blijkt dat de scoren aan de lage kant liggen en de vlinders vaak de verkeerde categorie krijgen opgelegd waardoor de ImageDataGenerator niet goed bijdraagt aan de generalisatie van het model.

```python
img = test_butterly_prediction(augmentation_model, butterfly_list[0])
print("afbeelding:")
img
```

Onderstaande gegevens tonen de score van het eerste model in vergelijking met het augumented model.
Het eerste model scoort 83.36% tegenover de 32.75% voor het model met data augmentatie.

```python
score = model.evaluate(val_ds, verbose=0)
print("Model zonder regularisatie")
print(f'Test loss: {score[0]} / Test accuracy: {score[1]}')

print()

score_augmentation = augmentation_model.evaluate(val_ds, verbose=0)
print("Augmentated model")
print(f'Test loss: {score_augmentation[0]} / Test accuracy: {score_augmentation[1]}')
```

## Evalueren van het model na toepassen van regularisatie

De 2 onderstaande functies printen de accuracy en loss waarden van beide modellen voor elke epoch.

```python
def plot_history(val1, val2, plot_name):
    c1 = history.history[val1]
    c2 = history.history[val2]
    
    epochs = range(len(c1))
  
    plt.plot(epochs, c1, 'b', label=val1)
    plt.plot(epochs, c2, 'r', label=val2)
    plt.title(str(plot_name))
    plt.legend(['train', 'test'], loc='upper left')
```

Op beide grafieken is te zien dat ze vrij evenredig bewegen. De accuracy beweegt omhoog en de beide loss waarden dalen steeds verder tot 0.6.

```python
plot_history('accuracy', 'val_accuracy', 'Accuracy vs Validation Accuracy')
```

```python
plot_history('loss', 'val_loss', 'Loss vs Validation Loss')
```

```python
def plot_augumented_history(val1, val2, plot_name):
    c1 = augmentation_history.history[val1]
    c2 = augmentation_history.history[val2]
    
    epochs = range(len(c1))
  
    plt.plot(epochs, c1, 'b', label=val1)
    plt.plot(epochs, c2, 'r', label=val2)
    plt.title(str(plot_name))
    plt.legend(['train', 'test'], loc='upper left')
```

Zoals je hieronder kunt zien, nemen het verlies aan training en validatie hier beide weinig toe in vergelijking met de uitkomst van het vorige model. Merk ook op hoe de nauwkeurigheid van training en validatie afneemt tegenover het vorige model. Ze zijn relatief dichterbij maar het model zonder augmentatie regulariseerde beter.
De loss van de validatieset (testset) is steeds hoger dan die van de trainingset terwijl de trainingset daalt, wat betekend dat de testset overfit is.

De accuracy waarden bewegen vrij gelijk mee met elkaar, de accuracy voor de testset ligt steeds dicht tegen de 20-30% aan wat vrij lage waarden zijn om het goed te doen op alle klasses.
